```python
everyinfoaboutschool = pd.concat([extracted_directory, extracted_performance], axis=1, sort=False)
everyinfoaboutschool["has_data"] = "True" 
everyinfoaboutschool.head()
```

```python
everyinfoaboutschool.rename(columns = {"dbn": "DBN"}, inplace = True)
everyinfoaboutschool.rename(columns = {"school_name": "School Name"}, inplace = True)
everyinfoaboutschool.rename(columns = {"boro": "Borough"}, inplace = True)
everyinfoaboutschool.rename(columns = {"zip": "zipcode"}, inplace = True)
everyinfoaboutschool.rename(columns = {"graduation_rate_2014": "highschool_graduation_rate"}, inplace = True)
everyinfoaboutschool.rename(columns = {"graduation_rate_boro": "Borough_Highschool_graduation_rate"}, inplace = True)
everyinfoaboutschool.rename(columns = {"college_career_rate_2014": "College Enrollment Rate"}, inplace = True)
everyinfoaboutschool.rename(columns = {"college_career_rate_boro": "Borough College Enrollment Rate"}, inplace = True)
everyinfoaboutschool.rename(columns = {"primary_address_line_1": "Primary_Address_Line_1"}, inplace = True)
everyinfoaboutschool.rename(columns = {"city": "City"}, inplace = True)
everyinfoaboutschool.rename(columns = {"state_code": "State_Code"}, inplace = True)
everyinfoaboutschool.head()
```

```python
has_data = {"Bronx": 0, "Brooklyn": 0, "Manhattan": 0, "Queens": 0, "Staten Island": 0}

for i, row in everyinfoaboutschool.iterrows():
    if type(row["highschool_graduation_rate"]) == float:
            if row["Borough"] == "Bronx":
                has_data["Bronx"] += 1
            elif row["Borough"] == "Brooklyn":
                has_data["Brooklyn"] += 1
            elif row["Borough"] == "Manhattan":
                has_data["Manhattan"] += 1
            elif row["Borough"] == "Queens":
                has_data["Queens"] += 1
            elif row["Borough"] == "Staten Island":
                has_data["Staten Island"] += 1

            everyinfoaboutschool.loc[i, "has_data"] = "False"
print(has_data)

total_has_data = 0

for Borough in has_data.values():
    total_has_data += Borough

print("Total deleted schools: {}".format(total_has_data))

new_everyinfoaboutschool = everyinfoaboutschool[everyinfoaboutschool.has_data!= "False"]
new_everyinfoaboutschool
```

ML Algorithm: Method 1 - Extreme Gradient Boosting
</p>
<p style="color:green;font-size:20px;"> 
Q: What assumptions need to be verified at this step?
</p>
<p style="color:green;font-size:20px;"> 
A: If there is relationship with all graduation rate and median income    
</p>
<p style="color:green;font-size:20px;"> 
Q: What algorithm did you choose?
</p>
<p style="color:green;font-size:20px;"> 
A: Extreme Gradient Boosting (xgboost)
</p>
<p style="color:green;font-size:20px;"> 
Q: Why you chose this algorithm?
</p>
<p style="color:green;font-size:20px;"> 
A: We ran our data against 17 models and Extreme Gradient Boosting would provide the 2nd best initial results.
</p>
<p style="color:green;font-size:20px;"> 
Q: What metric can measure the quality of your algorithm?
</p>
<p style="color:green;font-size:20px;"> 
A: During initial comparison against different models we used Mean Absolute Error (MAE), Mean Square Error (MSE), Root Mean Square Error (RMSE), R-squared (R2), Root Mean Squared Log Error (RMSLE), Mean Absolute Percentage Error (MAPE) as our metric. 
</p>
<p style="color:green;font-size:20px;"> 
Q: What is the outcome?
</p>
<p style="color:green;font-size:20px;"> 
A: We used 70% of the data on training the model. 20% of the data was used for test data and validation. Furthermore, we used 10% of data as unseen data to evaluate final results produced by the Extreme Gradient Boosting algorithm. The main parameter used for evaluation was RMSE, R2, and MAPE. We chose RMSE to achieve a well-balanced model. We chose R2 since it directly measures the goodness of fit in capturing the variance in training data. Additionally, we picked MAPE because it helps us understand the accuracy of the results. We tuned our parameter RMSE here to produce better results. We didn't tune R2 or MAPE since it would overfit the results. Essentially, our algorithm attempts to predict our target variable median income by merging estimates from a set of simpler and weaker models. 
</p>
<p style="color:green;font-size:20px;"> 
Q: Analysis and discussion: 
</p>
<p style="color:green;font-size:20px;"> 
A: Our initial RMSE value was 20400.0885, R2 value was 0.7559 and MAPE value was 0.1067. After tuning and optimizing, our RMSE value was recorded as 17562.1210, R2 was 0.9148, and our MAPE value was recorded as 0.0734 from the test data. Our final RMSE value for the unseen data was 255050868.1769, R2 value was 0.9237, and our MAPE value was 0.0847. 
    <br>
        <br>
Do these values provide an answer to our initial hypothesis? Yes, they show that our hypothesis was not entirely correct as there is no direct relationship between graduation rates and median income. Also, we noticed that when we used graduation rate as our only variable, our model failed to yield insightful results due to the very low value of R2. We had to support it with more variables such as Borough, Zip code, its relative Borough high school graduation rate and college enrollment rate to predict median income. 
    <br>
        <br>
Our most important feature plot here was centered on the Borough of Manhattan. Manhattan also had the third-highest number of high schools among the Boroughs featured in our data. Considering our prediction on test data and our plot for R2 below, our algorithm performed relatively well at predicting median income due to its high R2 value. These results leave room for further testing with our next algorithm, the Gradient Boosting Regressor (GBR). 
</p>
